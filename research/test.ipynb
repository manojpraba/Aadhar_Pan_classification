```python
def get_values_pan(img):
    # Initialize variables to None
    regex_name = None
    regex_dob = None
    regex_pan_number = None
    
    # Load SpaCy model for Name Entity Recognition
    NER = spacy.load("en_core_web_sm")
    
    # Convert image to grayscale
    thresh = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Configuration for name extraction
    img2str_config_name = "--psm 4 --oem 3"
    res_string_name = pytesseract.image_to_string(thresh, lang='eng', config=img2str_config_name)
    name_entities = NER(res_string_name)
    
    # Extract name using NER
    for word in name_entities.ents:
        if word.label_ == "PERSON":
            regex_name = re.findall("[A-Z ]+", word.text)
            print("Name (NER):", regex_name)
              # Stop after finding the first name entity
    
    # If name is not found using NER, attempt to extract from the string directly
    if not regex_name:
        regex_name = re.findall("[A-Z ]+", res_string_name)
        print("Name (OCR):", regex_name)
    
    # Configuration for extracting other information
    img2str_config_else = "--psm 3 --oem 3"
    res_string_else = pytesseract.image_to_string(thresh, lang='eng', config=img2str_config_else)
    
    # Extract date of birth
    regex_dob = re.findall("\d{2}/\d{2}/\d{4}", res_string_else)
    if regex_dob:
        regex_dob = regex_dob[0]
    else:
        # If date of birth is not found, try alternative pattern
        regex_dob = re.findall("\d{4}", res_string_else)
        if regex_dob:
            regex_dob = regex_dob[0]
    print("Date of Birth:", regex_dob)
    
    # Extract PAN number
    regex_pan_number = re.findall("[A-Z]{5}[0-9]{4}[A-Z]", res_string_else)
    if regex_pan_number:
        regex_pan_number = regex_pan_number[0]
    print("PAN Number:", regex_pan_number)

    return regex_name, regex_dob, regex_pan_number
```
